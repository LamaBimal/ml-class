```python
from sklearn.naive_bayes import GaussianNB
```

```python
#one sample
sample = np.array([4.1, 3.2, 1.2, 0.2]).reshape(1, -1) #reshape to one row - sample 
nb.predict_proba(sample).round(2)
```

## Support Vector Machine

```python
from sklearn.svm import LinearSVC, SVC
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions
from sklearn.metrics import accuracy_score
```

```python
from sklearn.datasets import make_circles
```

```python
#returns 2D array with data and corresponding class 
X, y =  make_circles(n_samples=1000, noise=0.2, factor=0.15)
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
plt.show()
```

### Linear Support Vector Machine Classifier

```python
lin_svc = LinearSVC()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
lin_svc.fit(X_train, y_train)

plot_decision_regions(X_train, y_train, lin_svc)
plt.show()
```

```python
y_pred = lin_svc.predict(X_test)
accuracy_score(y_test, y_pred)
```

## Using Kernel Functions

#### Degree 2 ploynomial kernel

```python
svc = SVC(kernel="poly", degree=2)

svc.fit(X_train, y_train)
plt.figure(figsize=(12,8))
plot_decision_regions(X_train, y_train, svc)
plt.show()
```

```python
y_pred = svc.predict(X_test)
accuracy_score(y_test, y_pred)
```

#### Degree 3 Ploynomial Kernel

```python
svc = SVC(kernel="poly", degree=3)

svc.fit(X_train, y_train)
plt.figure(figsize=(12,8))
plot_decision_regions(X_train, y_train, svc)
plt.show()
```

```python
y_pred = svc.predict(X_test)
accuracy_score(y_test, y_pred)
```

#### General Purpose RBF Kernel

```python
svc = SVC(kernel="rbf", C=1)
svc.fit(X_train, y_train)

plt.figure(figsize=(12,8))
plot_decision_regions(X_train, y_train, svc)
plt.show()
```

```python
y_pred = svc.predict(X_test)
accuracy_score(y_test, y_pred)
```

#### Increasing C Hyperparameter - Overfitting tendency increases

```python
svc = SVC(kernel="rbf", C=100)

svc.fit(X_train, y_train)
plt.figure(figsize=(12,8))
plot_decision_regions(X_train, y_train, svc)
plt.show()
```

```python
y_pred = svc.predict(X_test)
accuracy_score(y_test, y_pred)
```

#### Increasing Gamma -> Influence of inner points on boundary decreases - Only nearest poinst are considered as support vectors -> Overftting tendency increases

```python
svc = SVC(kernel="rbf", C=100, gamma=10)

svc.fit(X_train, y_train)
plt.figure(figsize=(12,8))
plot_decision_regions(X_train, y_train, svc)
plt.show()
```

```python
y_pred = svc.predict(X_test)
accuracy_score(y_test, y_pred)
```
